```python
import pandas as pd

df=pd.read_csv('car_price_prediction.csv')
df
```

```python
df["Levy"] = df["Levy"].str.replace("-","0")
df['Levy'] = df['Levy'].astype(int)
df
```

```python
df["Mileage"] = df["Mileage"].str.replace(" km","")
df['Mileage'] = df['Mileage'].astype(int)
df
```

```python
df = df.loc[(df['Price'] > 1000) & (df['Price'] < 1000000)]
df
```

```python
df['Prod. year'] = pd.to_datetime(df['Prod. year'], format='%Y')
df['Prod. year'] = df['Prod. year'].dt.year
df
```

```python
df = df.drop_duplicates()
df
```

```python
if df['ID'].duplicated().any():
    print("Il y a des doublons dans la colonne 'ID'")
else:
    print("Il n'y a pas de doublons dans la colonne 'ID'")
```

```python
duplicates = df[df.duplicated(['ID'], keep=False)]['ID']
duplicates.value_counts()
```
